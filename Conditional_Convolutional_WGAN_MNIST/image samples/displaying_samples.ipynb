```python
import torch
import matplotlib.pyplot as plt
```

```python
parent_dir = 'sample_with_different_lr/'
img1 = torch.load(f'{parent_dir}generated_img_sample_epoch_10_lr(0.001,0.001).pt')
img2 = torch.load(f'{parent_dir}generated_img_sample_epoch_10_lr(0.001,0.0005).pt')
img3 = torch.load(f'{parent_dir}generated_img_sample_epoch_10_lr(0.001,0.0001).pt')

num = 9

fig,axes = plt.subplots(1,3,figsize=(10,30))
axes = axes.flatten()
for axe in axes:
    axe.xaxis.set_ticklabels([])
    axe.yaxis.set_ticklabels([])

axes[0].set_xlabel("critic's lr: 0.001\ngenerator's lr: 0.001",size=10)
axes[1].set_xlabel("critic's lr: 0.001\ngenerator's lr: 0.0005",size=10)
axes[2].set_xlabel("critic's lr: 0.001\ngenerator's lr: 0.0001",size=10)
axes[0].imshow(img1[num].permute(1,2,0),'gray')
axes[1].imshow(img2[num].permute(1,2,0),'gray')
axes[2].imshow(img3[num].permute(1,2,0),'gray')
fig.suptitle('Generated Samples With Different Learning Rates in Epoch 10\nmodel is trying to make handwritten number 9'
             ,size=10,y=0.547,fontweight='bold')
# plt.tight_layout()
# plt.subplots_adjust(top=0.9,hspace=0.001)  # Increase top margin
plt.show()
```

```python
img_num = 10
imgs_dict = {}
folder_dir = 'samples_during_training/iteration_2/'
fig,axes = plt.subplots(img_num,10,figsize=(40,img_num*4))
axes = axes.flatten()
for i in range(0,img_num*10):
    if (i)%10 == 0:
        img_dir = f'{folder_dir}generated_img_sample_epoch_{i+10}.pt'  
        img = torch.load(img_dir)
    if i%10 == 0:
        axes[i].set_ylabel(f'epoch {i+10}',size=40)
    if i>89: 
        axes[i].set_xlabel(f'{i%10}',loc='center',size=70)
        
    axes[i].xaxis.set_ticklabels([])
    axes[i].yaxis.set_ticklabels([])
    axes[i].imshow(img[i%10].permute(1,2,0),'gray')

fig.suptitle('Generated Samples during Training\n every 10 epochs'
             ,fontsize=50,fontweight='bold',y=0.994)
plt.tight_layout()
fig.subplots_adjust(top=0.95)
plt.show()
```
