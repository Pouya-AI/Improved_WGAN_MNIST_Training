```python
import torch
import torch.nn as nn
import torch.optim as optim 
import matplotlib.pyplot as plt
from Generator import Generator
```

```python
gen = Generator('cpu').eval()
state_dict = torch.load('generator_and_critic/Unconditional_Convolutional_Generator_state_dict.pth')
gen.load_state_dict(state_dict)
```

```python
z = torch.randn(10,32)
with torch.no_grad():
    imgs = gen(z)
fig,axes = plt.subplots(1,10,figsize=(10,1))
for i in range(len(axes)):
    axes[i].axis('off')
    axes[i].imshow(imgs[i].permute(1,2,0),'gray')
```
